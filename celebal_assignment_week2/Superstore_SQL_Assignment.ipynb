# Step 1: Import Required Libraries

```python
import pandas as pd
import duckdb
```

# Step 2: Load the Dataset

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
df.head()
```

# Step 3: Load Dataset into SQL Database

```python
con = duckdb.connect()
con.register("superstore", df)
```

# Step 4: Explore the Table

```python
con.sql("""
SELECT *
FROM superstore
LIMIT 5
""").df()
```

#Total Rows

```python
con.sql("""
SELECT COUNT(*) AS Total_Rows
FROM superstore
""").df()
```

#Column Names

```python
df.columns
```

#Data Types

```python
df.dtypes
```

Apply WHERE Filters

```python

#Region = West
con.sql("""
SELECT *
FROM superstore
WHERE Region='West'
""").df()


```

```python
#Category = Furniture
con.sql("""
SELECT *
FROM superstore
WHERE Category='Furniture'
""").df()
```

```python
#Sales > 500
con.sql("""
SELECT *
FROM superstore
WHERE Sales>500
""").df()
```

 GROUP BY Queries

```python
con.sql("""
SELECT Category,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
""").df()
```

```python
#Quantity by Region
con.sql("""
SELECT Region,
SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Region
""").df()
```

```python
#Average Profit
con.sql("""
SELECT Category,
AVG(Profit) AS Average_Profit
FROM superstore
GROUP BY Category
""").df()
```

 ORDER BY and LIMIT

```python
con.sql("""
SELECT "Product Name",
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10
""").df()
```

```python
#Top Categories
con.sql("""
SELECT Category,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC
""").df()
```

Business Queries

```python
#Monthly Sales Trend

con.sql("""
SELECT strftime('%Y-%m', CAST("Order Date" AS DATE)) AS Month,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Month
ORDER BY Month
""").df()
```

```python
#Monthly Sales Trend

con.sql("""
SELECT strftime('%Y-%m', strptime("Order Date", '%m/%d/%Y')) AS Month,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Month
ORDER BY Month
""").df()
```

```python
#Top Customers

con.sql("""
SELECT "Customer Name",
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10
""").df()
```

```python
#Duplicate Records
con.sql("""
SELECT *,
COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY ALL
HAVING COUNT(*)>1
""").df()
```
